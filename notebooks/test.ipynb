```python
from pyPTV.pyPTV import PTV
import mdtraj as md
```

```python
traj = md.load(
    "data/ala5.dcd",
    top="data/ala5.pdb",
)
```

```python
v = PTV(
    name="ala5",
    traj=traj,
    step=(20.0, "ns"),
)
```

```python
v.plot_dssp()
```
